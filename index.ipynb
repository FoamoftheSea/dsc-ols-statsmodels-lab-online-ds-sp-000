## Step 1: Read the dataset and inspect its columns and 5-point statistics

```python
# Load necessary libraries and import the data
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')
```

```python
# Check the columns and first few rows
df = pd.read_csv('advertising.csv', index_col=0)
df.head()
```

```python
# Get the 5-point statistics for data 
df.describe()
```

```python
df.info()
```

```python
# Describe the contents of this dataset
# It looks like the data has an index that starts at 1 in the column "Unnamed: 0"
# The remaining columns are the spending per ad type (TV, radio, newspaper), and the sales
# At first glance using .info() there does not appear to be any missing data
```

## Step 2: Plot histograms with kde overlay to check the distribution of the predictors

```python
# For all the variables, check distribution by creating a histogram with kde
for column in df:
    df[column].plot.hist(density=True, label = column+' histogram', alpha=0.8);
    df[column].plot.kde(label= column+' kde');
    plt.legend();
    plt.show();
```

```python
df.boxplot(grid=True, figsize=(12,8))
```

```python
# Record your observations here 
# None of the variables are normally distributed, but have some similarities in that they have
# centralized means and few outliers.
# The newspaper column is heavily positively skewed
# The sales column has a slight positive skew, almost normally distributed
# There are only outliers in the newspaper column
```


## Step 3: Test for the linearity assumption 

Use scatterplots to plot each predictor against the target variable

```python
# visualize the relationship between the preditors and the target using scatterplots
fig, ax = plt.subplots(1, 3, sharey=True, figsize=(12,4))
for index, channel in enumerate(['TV','radio','newspaper']):
    print(index, channel)
    df.plot.scatter(x=channel, y='sales', ax=ax[index], label=channel)
plt.legend()
plt.show()
```

```python
# Record yor observations on linearity here 
# There appears to be some linearity for TV and radio, but not so much for newspaper
# TV is more linear than radio
# Though there is linearity in TV and radio, it looks like there may be heteroscedasticity
# present after fitting a linear model becuase of the cone like shapes
```

### Conclusion so far

Based on above initial checks, we can confidently say that TV and radio appear to be good predictors for our regression analysis. Newspaper is very heavily skewed and also doesnt show any clear linear relationship with the target. 
> We'll move ahead with our analysis using TV and radio, and rule out newspaper because we believe it violates OLS assumptions

Note: Kurtosis can be dealt with using techniques like log normalization to "push" the peak towards the center of distribution. You'll learn about this later on.

## Step 4: Run a simple regression in Statsmodels with TV as a predictor

```python
# build the formula 
f = 'sales~TV'
# create a fitted model in one line
model = ols(formula=f, data=df).fit()
```

## Step 5: Get Regression Diagnostics Summary

```python
model.summary()
```

```python
# R-squared value is 0.61, not a great fit. 61% of variance in sales can be explained using the
# spendings on TV
# Each unit increase of TV spending is associated with a increase of 0.0475 units of sales
# This means that for every 100 dollars spent in TV ads, there is an associated $4.75 gain of sales
```

Note here that the coefficients represent associations, not causations

## Step 6:  Draw a prediction line with data points on a scatter plot for X (TV) and Y (Sales)

Hint: You can use the `model.predict()` function to predict the start and end point of of regression line for the minimum and maximum values in the 'TV' variable. 

```python
# create a DataFrame with the minimum and maximum values of TV
X_new = pd.DataFrame({'TV':[df.TV.min(), df.TV.max()]})
print(X_new.head())
# make predictions for those x values and store them
predicts = model.predict(X_new)
print(predicts)

# first, plot the observed data and the least squares line
df.plot.scatter(x='TV', y='sales', label="Data")
plt.plot(X_new, predicts, c='r', label="Regression Line")
plt.legend()
plt.show()
```

## Step 7: Visualize the error term for variance and heteroscedasticity

```python
fig = plt.figure(figsize=(12,8))
fig = sm.graphics.plot_regress_exog(model,"TV",fig=fig)
plt.show()
```

```python
# Record Your observations on heteroscedasticity
# As anticipated, there is heteroscedasticity in the data, as the scatter plot of the residuals
# creates a cone shape, meaning that the variability/scatter of the data is not consistent over
# the predictor values
```

## Step 8: Check the normality assumptions by creating a QQ-plot

```python
# Code for QQ-plot here
import scipy.stats as stats
residuals = model.resid
fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45',fit=True)
fig.show()
```

```python
# Record Your observations on the normality assumption
# The Q-Q plot shows that the residuals are normally distributed, meaning the Normality assumption
# can stand
```

## Step 9: Repeat the above for radio and record your observations

```python
# code for model, prediction line plot, heteroscedasticity check and QQ normality check here
f2 = 'sales~radio'
model2 = ols(formula=f2, data=df).fit()

X_new = pd.DataFrame({'radio':[df.radio.min(), df.radio.max()]})
predicts = model2.predict(X_new)

df.plot.scatter(x='radio', y='sales', label="Data")
plt.plot(X_new, predicts, c='r', label="Regression Line")
plt.legend()
plt.show()

fig = plt.figure(figsize=(12,8))
fig = sm.graphics.plot_regress_exog(model2,"radio",fig=fig)
plt.show()

residuals = model2.resid
fig2 = sm.graphics.qqplot(residuals, dist=stats.norm, line='45',fit=True)
fig2.show()
```

```python
model2.summary()
```

```python
# Record your observations here for goodness of fit 
# The R_squared value is low (0.332), meaning the fit is not great, as only 32% of the variance
# in sales can be explained using the spendings on radio
# The residuals QQ plot seems a bit too curved to assume normality
# However, the cone shape of the scatter plot of residuals shows that there is heteroscedasticity
# in the model
```
